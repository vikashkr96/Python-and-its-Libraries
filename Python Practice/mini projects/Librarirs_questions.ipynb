```python
# a) Usage of methods such as floor ( ), ceil ( ), sqrt ( ), isqrt ( ), gcd ( ) etc.

import math

marks = 87.6

section_A = 36
section_B = 48

floor_value = math.floor(marks)
ceil_value = math.ceil(marks)

sq_root = math.sqrt(marks)
int_sq_root = math.isqrt(floor_value)

max_groups = math.gcd(section_A, section_B)


print("Original Marks:", marks)
print("Floor Marks:", floor_value)
print("Ceil Marks:", ceil_value)
print("Square Root of Marks:", sq_root)
print("Integer square root:", int_sq_root)
print("Maximum Equal Groups:", max_groups)
```

```python
# b) Usage of attributes of array such as ndim, shape, size, methods such as sum ( ), mean ( ), sort ( ), sin ( ) etc.

import numpy as np

arr = np.array([[1,2,3],
                [7,10,0],
                 [4,5,6]])

dimension = arr.ndim
shape = arr.shape
size = arr.size
total = np.sum(arr)
avg = np.mean(arr)
sorted_arr = np.sort(arr)  # row wise sorting
sorted_arr1 = np.sort(arr, axis=0)  # column wise sorting
sine_of_elem = np.sin(arr)

print("Dimension of the array arr is:", dimension)
print("Shape of the array arr is:", shape)
print("Size of the array arr is:", size)
print("Sum of the elements of array arr is:", total)
print("Average of the elements array arr is:", avg)
print("Row wise sorted arr is:\n", sorted_arr)
print("Column wise sorted arr is:\n", sorted_arr1)
print("Sine of Each element of arr is:\n", sine_of_elem)
```

```python
# c) Usage of methods such as det ( ), eig ( ) etc.

arr1 = np.array([[1,2],
                [3,4]])

det_value = np.linalg.det(arr1)
eig_val = np.linalg.eig(arr1)

print("Determinant Value of arr1 is:", det_value)
print("Eigen values and Vectors are\n:", eig_val)
```

```python
# d) Consider a list data type (1D) then reshape it into 2D, 3D matrix using numpy.

arr2 = np.array([1,2,3,4,5,6])

two_D  = arr2.reshape(2,3)
three_D = arr2.reshape(1,2,3)

print("Two Dimensional Form of Array arr2 is:\n", two_D)
print("Three Dimensional Form of Array arr2 is:\n", three_D)
```

```python
# e) Generate random matrices using numpy.

random_mat = np.random.rand(2,3)
random_mat1 = np.random.rand(5)

print("Random Matrices are:\n", random_mat, "\n And \n", random_mat1)
```

```python
# f)  Find the determinant of a matrix using scipy.
# g)  Find eigen value and eigenvector of a matrix using scipy.

import numpy as np
from scipy import linalg

A = np.array([[1, 2],
              [3, 4]])

det_A = linalg.det(A)
eigen_values, eigen_vectors = linalg.eig(A)

print("Determinant:", det_A)
print("Eigen Values:\n", eigen_values)
print("Eigen Vectors:\n", eigen_vectors)
```
