# **Advance Operation With Business Examples**

```python
import numpy as np
import matplotlib.pyplot as plt


# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])


# Total sales (yearly)...
print("Yearly Sales : ", np.sum(sales_data[:,1:], axis=0)) # to add all rows in a column

# Total sales in each restaurants...
print("Sales in each restorants :", np.sum(sales_data[:,1:],axis = 1))

# Minimum sales in a year ..
min_sales = np.min(sales_data[:,1:],axis= 0)
print(min_sales)
print("Minimum sale in a year is: ", min(min_sales))

# yearly average sales per restaurant

avg_sales = np.mean(sales_data[0:,1:], axis = 1)
print("yearly average sales per restorant is: ", avg_sales)

# yearly average sales per restaurant (2nd method)

avg_sales_ = np.sum(sales_data[:, 1:], axis=1) / sales_data[:, 1:].shape[1]
print(avg_sales_)

# cummulative sales per restaurant
print("\n Cummulative sales per restaurant:")
cumsum = np.cumsum(sales_data[:,1:])
print(cumsum)

```

# **Vectorized Operations**

```python
import numpy as np

vector1 = np.array([[1,2,3,4],[5,6,7,8]])
vector2 = np.array([[9,10,11,12],[13,14,15,16]])

vector_sum = (vector1 + vector2)
vector_diff = (vector2 - vector1)
vector_mul = (vector1 * vector2)

print(f" Vector Sum is: {vector_sum},\n Vector Difference is: {vector_diff}\n And Vector Multiplication is: {vector_mul}")

# dot product
vector3 = np.arange(0,8)
temp = vector3.reshape(4,2)
print(" Dot Product is: ", np.dot(vector1,temp))
```

```python
# Angle bt two Vectors (formula => cosine of angle = a dot b upon mag of a and b)

angle_in_deg = np.arccos((np.dot(vector1,temp))/((np.linalg.norm(vector1))*(np.linalg.norm(temp))))  # "np.linalg.norm(vector1)"   is basically the magnitude or normalization of vector 1

print(angle_in_deg)
```
