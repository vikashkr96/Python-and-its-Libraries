# loading numpy logo and showing it through matplotlib

```python
try:
  logo = np.load("numpy_logo.npy")
  # display the logo
  plt.figure(figsize=(10,5))
  plt.subplot(121)
  plt.imshow(logo)
  plt.title("Numpy Logo")
  plt.grid(False)
  plt.show()

# converting into dark (inverting the logo)
  dark_logo = 1 - logo
  plt.figure(figsize=(10,5))
  plt.subplot(121)
  plt.imshow(dark_logo)
  plt.title("Numpy Dark Logo")
  plt.grid(False)

except FileNotFoundError:
  print("Numpy logo does not found!")


# basically its  also uses the concept of broadcasting , all pixels of the image are stored in a numpy array bt 0 and 1
# 0 is black and 1 is white => and values bt them are the coloured once
# by doing 1 - logo (broadcasting used) , it actually converts all the pixel into negatives color pixels and hence the color is inverted
```

```python
# converting my image with png or jpg to npy extension and then storing it into numpy arrays
# showing it and also its dark version
```

```python
from PIL import Image
import numpy as np

img = Image.open('me.jpg')
# Convertin the image to a NumPy array and saving it in npy file
img_np_array = np.array(img)
np.save("img_np_array.npy",img_np_array)

mypic = np.load("img_np_array.npy")
# display the pic
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(mypic)
plt.title("Its Me")
plt.grid(False)
plt.show()

#  converting into negatrive /dark
my_darkpic = 1 - mypic
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(my_darkpic)
plt.title("Its My dark side")
plt.grid(False)
plt.show()
```
